```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization,Flatten
from keras.optimizers import Adam
%matplotlib inline
```

```python
dataset=pd.read_csv('dataset/titanic/train.csv')
dataset.head()
```

```python
Y=dataset['Survived']
X=dataset
X=X.drop("Ticket",axis=1)#ticket is alpha numeric and non interpretable
X=X.drop("Pclass",axis=1)
X=X.drop("Cabin",axis=1)#cabin is removed because it is alpha numeric and non interpretable
X=X[X.Embarked=='S']#considering only people who have embarked on ship


model=Sequential()
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(2,activation='softmax',kernel_initializer='random_uniform'))
model.compile(loss='mean_squared_error',
              optimizer=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False),
             metrics=['accuracy'])
```
