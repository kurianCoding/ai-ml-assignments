```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('dataset/suicide_rate_overview/master.csv')
unique_country=dataset['country'].unique()
alpha=0.8
plt.figure(figsize=(10,25))
sns.countplot(y='country',data=dataset,alpha=alpha,palette='viridis')
plt.show()
```

```python
sns.heatmap(dataset.corr(),cmap='bone',annot=True)
```
